## Seaborn 시각화

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams["font.family"] = 'AppleGothic'
plt.rcParams['font.size'] = 10
mpl.rcParams['axes.unicode_minus'] = False
```

```python
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

- countplot()

```python
sns.countplot(x = 'class', data = titanic)
plt.title('Passengers of titanic by class')
plt.show()
```

```python
sns.countplot(x="day", data=tips)
plt.title("요일별 팁을 준 횟수")
plt.show()
```

- pairplot()

```python
sns.pairplot(iris)
plt.show()
```

```python
sns.pairplot(iris, hue="species", markers=["o", "s", "D"])
plt.title("Iris Pair Plot, Hue로 꽃의 종을 시각화")
plt.show()
```

- heatmap()

```python
flights_passengers = flights.pivot("month", "year", "passengers")
sns.heatmap(flights_passengers, annot = True, fmt = 'd')
plt.show()
```

- barplot()

```python
sns.barplot(x="day", y="total_bill", data=tips)
plt.title("요일 별, 전체 요금")
plt.show()
```

```python
sns.barplot(x="day", y="total_bill", hue = 'sex', data=tips)
plt.title("요일 별/성별 전체 요금")
plt.show()
```

- boxplot()

```python
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("요일 별 전체 요금의 Box Plot")
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", hue = 'sex', data=tips)
plt.title("요일 별/성별 전체 요금의 Box Plot")
plt.show()
```
