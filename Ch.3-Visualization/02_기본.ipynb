## 데이터 시각화

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["font.family"] = 'AppleGothic'
plt.rcParams['font.size'] = 10
mpl.rcParams['axes.unicode_minus'] = False
```

### 1. 라인 플롯

```python
plt.plot([1, 4, 9, 16])
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.title('Squared')
plt.grid()
plt.show() 
```

- 스타일 지정

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'rs--')
plt.title('rs-- style')
plt.grid()
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], c = 'purple', lw = 3, marker = 'o', ms = 8)
plt.title('colour, line width, marker size')
plt.grid()
plt.show()
```

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c="b",
         lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.title("스타일 적용 예")
plt.show()
```

- 그림 범위 지정

```python
plt.plot([1, 2, 3, 4,], [1, 4, 9, 16], 'c+:', ms = 8)
plt.xlim(0, 5)
plt.ylim(-2, 20)
plt.grid()
plt.show()
```

- 틱 설정

```python
x = np.linspace(-np.pi, np.pi, 300)
y = np.cos(x)
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 300)
y = np.cos(x)
plt.plot(x, y)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
plt.yticks([-1, -0.5, 0, .5, 1])
plt.grid()
plt.title('그래프의 위치 표시를 틱이라고 한다')
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 300)
y = np.cos(x)
plt.plot(x, y)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.grid()
plt.title('틱은 문자도 사용이 된단다')
plt.show()
```

- 여러개의 선 그리기

```python
t = np.arange(0., 5., 0.2)
plt.title("라인 플롯에서 여러개의 선 그리기")
plt.plot(t, t, 'r--', t, 0.5 * t**2, 'bs:', t, 0.2 * t**3, 'g^-')
plt.grid()
plt.show()
```

- plot함수를 여러번 사용해 겹쳐 그리기

```python
plt.plot(t, t, 'r--')
plt.plot(t, 0.5 * t**2, 'bs:')
plt.plot(t, 0.2 * t**3, 'g^-')
plt.grid()
plt.show()
```

- 범례(legend)

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("legend는 플롯을 설명해준데")
plt.plot(X, C, ls="--", label="cosine")
plt.plot(X, S, ls=":", label="sine")
plt.legend(loc=2)   # loc = int | 범례의 위치를 지정
plt.grid()
plt.show()
```

- x축, y축 라벨, 타이틀

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("X, Y 라벨링")
plt.plot(X, C, ls="--", label="cosine")
plt.plot(X, S, ls=":", label="sine")
plt.legend(loc=2)
plt.xlabel('X축'); plt.ylabel('Y축')
plt.grid()
plt.show()
```
