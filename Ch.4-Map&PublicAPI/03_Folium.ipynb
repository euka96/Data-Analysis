```python
map = folium.Map(location = [37.5598688323642, 126.967109652664], zoom_start = 14)
folium.Marker(location = [37.5598688323642, 126.967109652664], popup = '한국경제신문사').add_to(map)
folium.Marker(location = [37.5663174209601, 126.977829174031],
                tooltip = '서울시청', icon = folium.Icon(color = 'red', icon = 'info-sign')).add_to(map)
folium.Circle(location = [37.5641201543296, 126.998009728978],
                radius = 100, tooltip = '중구청', color = 'crimson', fill = False).add_to(map)
map
```

```python
df = pd.read_csv('addr2.csv')
df
```

```python
map = folium.Map(location = [df['latitude'].mean(), df['longitude'].mean()], zoom_start = 14)
for i in df.index :
    folium.Marker(location = [df['latitude'][i], df['longitude'][i]],
                    popup = folium.Popup(df['address'][i], min_width = 200, max_width = 200),   # 누르면 내용이 뜸
                        tooltip = df['name'][i]).add_to(map)
map_html = '<h3 align = "center" style = "font-size : 20px"> <b>5개 지역 표시</b> </h3>'
map.get_root().html.add_child(folium.Element(map_html))
map
```

```python
park = pd.read_csv('서울시 주요 공원현황.csv', encoding = 'euckr')
park = park[['Y좌표(WGS84)', 'X좌표(WGS84)','공원주소','공원명']].dropna(axis = 0)
seoul_park = folium.Map(location = [park['Y좌표(WGS84)'].mean(), park['X좌표(WGS84)'].mean()], zoom_start = 11)
for i in park.index :
    folium.Marker(location = [park['Y좌표(WGS84)'][i], park['X좌표(WGS84)'][i]],
                    popup = folium.Popup(park['공원주소'][i], min_width = 200, max_width = 200),
                        tooltip = park['공원명'][i]).add_to(seoul_park)
seoul_park
```
