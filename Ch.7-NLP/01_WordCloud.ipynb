```python
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
```

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams["font.family"] = 'AppleGothic'
plt.rcParams['font.size'] = 10
mpl.rcParams['axes.unicode_minus'] = False
```

## Alice in Wonderland

```python
with open('data/Alice.txt', 'r') as f :
    text = f.read()
stopwords = set(STOPWORDS)
len(stopwords)
```

```python
stopwords.add('said')
```

```python
wc = WordCloud(background_color = 'white', max_words = 2000, stopwords = stopwords)
wc = wc.generate(text)
```

```python
keys = list(wc.words_.keys())
values = list(wc.words_.values())

for i in range(10) :
    print(f'{keys[i]} : {values[i]}')
```

```python
plt.figure(figsize = (10,6))
plt.imshow(wc, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

### WordCloud on Photo

```python
mask = np.array(Image.open('data/Alice_mask.png'))
plt.figure(figsize = (10,6))
plt.imshow(mask, cmap = plt.cm.gray)   # cm.gray | 흑백으로 표시
plt.axis('off')
plt.show()
```

```python
wc = WordCloud(background_color = 'white', max_words = 2000, stopwords = stopwords, mask = mask)
wc = wc.generate(text)
```

```python
plt.figure(figsize = (10,10))
plt.imshow(wc, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

## Star Wars

```python
with open('data/A_new_hope.txt', 'r') as f:
    text = f.read()
text = text.replace("HAN", 'Han')
text = text.replace("LUKE'S", 'Luke')
mask = np.array(Image.open('data/Stormtrooper_mask.png'))
```

```python
stopwords = set(STOPWORDS)
stopwords.add('INT')
stopwords.add('ext')
```

```python
wc = WordCloud(max_words = 2000, stopwords = stopwords, mask = mask, margin = 10, random_state = 1).generate(text)
default_colour = wc.to_array()
```

```python
import random
def grey_colour(word, font_size, position, orientation, random_state = None, **kwargs) :
    #return 'hsl(0, 0%%, %d%%)' % random.randint(60, 100)    # hue, saturation, lightness
    return f'hsl(0, 0%, {random.randint(60, 100)}%)'
```

```python
plt.figure(figsize = (10,10))
plt.imshow(wc.recolor(color_func = grey_colour, random_state = 3), interpolation = 'bilinear')
plt.axis('off')
plt.show()
```
